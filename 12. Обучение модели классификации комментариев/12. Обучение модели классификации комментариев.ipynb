 # Проект для «Викишоп»

Интернет-магазин «Викишоп» запускает новый сервис. Теперь пользователи могут редактировать и дополнять описания товаров, как в вики-сообществах. То есть клиенты предлагают свои правки и комментируют изменения других. Магазину нужен инструмент, который будет искать токсичные комментарии и отправлять их на модерацию.

Обучите модель классифицировать комментарии на позитивные и негативные. В вашем распоряжении набор данных с разметкой о токсичности правок.

Постройте модель со значением метрики качества F1 не меньше 0.75.

**Описание данных**

Данные находятся в файле toxic_comments.csv. Столбец text в нём содержит текст комментария, а toxic — целевой признак.

```python
import numpy as np
import pandas as pd
import en_core_web_sm
import spacy
import re

import nltk
from nltk.corpus import stopwords as nltk_stopwords

from tqdm import tqdm, tqdm_notebook
tqdm_notebook().pandas()

from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.metrics import f1_score
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer

import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

# Подготовка данных

```python
nlp = en_core_web_sm.load()
"""Функция лемматизирует текст"""
def lemmatize_spacy(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])
```

```python
"""Функция очищает от лишних символов, пробелов и приводит к нижнему регистру"""
def clear_text(text):
    clean_list = re.sub(r'[^a-zA-Z]'," ", text.lower())
    clean_list = clean_list.split()
    clean_list = " ".join(clean_list)
    return clean_list
```

Применим функции лемматизации и очистки

```python
data['clean_text'] = data['text'].progress_apply(lemmatize_spacy)
```

```python
data['clean_text'] = data['clean_text'].progress_apply(clear_text)
```

```python
data.sample(7)
```

Как видно из облака чаще всего встречаются слова: article(статья),think(думать),make(делать),edit(редактировать),page(страница) и еще много других,которые действительно определяют тематику нашего задания и не носят негативный характер.

Из данного облака видно, что негативные комментарии размечены верно и какое слово является самым употребляемым )))

```python
# Разобьем данные на обучающую и тестовую выборку с учетом дисбаланса

train, test = train_test_split(data, test_size=0.25,random_state=12345,stratify=data['toxic'])
```

```python
print('Отношение целевого признака в обучающей выборке:',train['toxic'].value_counts()[1]/train['toxic'].value_counts()[0])
print('Отношение целевого признака в тестовой выборке:',test['toxic'].value_counts()[1]/test['toxic'].value_counts()[0])
```

```python
plt.figure(figsize=(6, 6))
plt.subplot(211)
train['toxic'].hist()
plt.title('Распределение целевого признака train')

plt.subplot(212)
test['toxic'].hist()
plt.title('Распределение целевого признака test')


```

```python
# Создали корпусы и отделили целевой признак в обеих выборках

corpus_train = list(train['clean_text']) # создали корпус из трейна
train_y = train['toxic']

corpus_test = list(test['clean_text']) # создали корпус из теста
test_y = test['toxic']
```

# Обучение и выбор лучшей модели 

```python
stopwords = set(nltk_stopwords.words('english'))# прогрузили стоп-слова на английском
```

Для выбора модели будем использовать pipline

```python
pipe = Pipeline(
    [
        ("vect", TfidfVectorizer(stop_words = stopwords,ngram_range=(1,2))),
        ("clf", LogisticRegression()),
    ]
)
```

Сначала без подбора гиперпараметров выберем модель. Где есть возможность,указываем на необходимость сбалансировать данные.

```python
%%time

param_grid = [
    {
        "clf": 
        [
        LogisticRegression(random_state=12345,class_weight='balanced'),
        SGDClassifier(random_state=12345,class_weight='balanced'),
        LinearSVC(class_weight='balanced', random_state=12345),
        ],
    }
]
grid = GridSearchCV(pipe, param_grid=param_grid, scoring='f1', cv=5, n_jobs=-1)
grid.fit(corpus_train, train_y)
```

```python
print(f'Лучшая F1:{grid.best_score_:.4f}')
```

```python
print('Лучшая модель:')
grid.best_estimator_.named_steps.clf
```

Подберем для лучшей модели гиперпараметры

```python
param_grid_best_model = [
    {
        "clf": [LinearSVC(class_weight='balanced', random_state=12345)],
        "clf__C" : (0.1,0.56,2.52,3.16),
       
    },
]
```

```python
%%time
grid_best_model = GridSearchCV(pipe, param_grid=param_grid_best_model, scoring="f1", cv=5,n_jobs=-1)
grid_best_model.fit(corpus_train, train_y)
```

```python
print(f'Метрика F1 на лучшей модели с подобранными гиперпараметрами: {grid_best_model.best_score_:.4f}')
```

```python
print('Лучшая модель')
grid_best_model.best_estimator_.named_steps.clf
```

## Тестирование лучшей модели

```python

print(f'F1 на тестовом наборе данных: {f1_score(test_y, grid_best_model.predict(corpus_test)):.4f}')
```
