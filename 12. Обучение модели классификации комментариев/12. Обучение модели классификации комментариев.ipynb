# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

```python
nlp = en_core_web_sm.load()
"""–§—É–Ω–∫—Ü–∏—è –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä—É–µ—Ç —Ç–µ–∫—Å—Ç"""
def lemmatize_spacy(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])
```

```python
"""–§—É–Ω–∫—Ü–∏—è –æ—á–∏—â–∞–µ—Ç –æ—Ç –ª–∏—à–Ω–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤, –ø—Ä–æ–±–µ–ª–æ–≤ –∏ –ø—Ä–∏–≤–æ–¥–∏—Ç –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É"""
def clear_text(text):
    clean_list = re.sub(r'[^a-zA-Z]'," ", text.lower())
    clean_list = clean_list.split()
    clean_list = " ".join(clean_list)
    return clean_list
```

–ü—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏–∏ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏–∏ –∏ –æ—á–∏—Å—Ç–∫–∏

```python
data['clean_text'] = data['text'].progress_apply(lemmatize_spacy)
```

```python
data['clean_text'] = data['clean_text'].progress_apply(clear_text)
```

```python
pipe = Pipeline(
    [
        ("vect", TfidfVectorizer(stop_words = stopwords,ngram_range=(1,2))),
        ("clf", LogisticRegression()),
    ]
)
```
