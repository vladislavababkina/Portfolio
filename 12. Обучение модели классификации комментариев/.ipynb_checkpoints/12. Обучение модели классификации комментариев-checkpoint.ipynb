–û–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ –Ω–∞ –ø–æ–∑–∏—Ç–∏–≤–Ω—ã–µ –∏ –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã–µ. –í –≤–∞—à–µ–º —Ä–∞—Å–ø–æ—Ä—è–∂–µ–Ω–∏–∏ –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö —Å —Ä–∞–∑–º–µ—Ç–∫–æ–π –æ —Ç–æ–∫—Å–∏—á–Ω–æ—Å—Ç–∏ –ø—Ä–∞–≤–æ–∫.

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –º–æ–¥–µ–ª—å —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –º–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ F1 –Ω–µ –º–µ–Ω—å—à–µ 0.75.

**–û–ø–∏—Å–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö**

–î–∞–Ω–Ω—ã–µ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ —Ñ–∞–π–ª–µ toxic_comments.csv. –°—Ç–æ–ª–±–µ—Ü text –≤ –Ω—ë–º —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç–µ–∫—Å—Ç –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è, –∞ toxic ‚Äî —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫.

```python
import numpy as np
import pandas as pd
import en_core_web_sm
import spacy
import re

import nltk
from nltk.corpus import stopwords as nltk_stopwords

from tqdm import tqdm, tqdm_notebook
tqdm_notebook().pandas()

from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.metrics import f1_score
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer

import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

```python
nlp = en_core_web_sm.load()
"""–§—É–Ω–∫—Ü–∏—è –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä—É–µ—Ç —Ç–µ–∫—Å—Ç"""
def lemmatize_spacy(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc])
```

```python
"""–§—É–Ω–∫—Ü–∏—è –æ—á–∏—â–∞–µ—Ç –æ—Ç –ª–∏—à–Ω–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤, –ø—Ä–æ–±–µ–ª–æ–≤ –∏ –ø—Ä–∏–≤–æ–¥–∏—Ç –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É"""
def clear_text(text):
    clean_list = re.sub(r'[^a-zA-Z]'," ", text.lower())
    clean_list = clean_list.split()
    clean_list = " ".join(clean_list)
    return clean_list
```

–ü—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏–∏ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏–∏ –∏ –æ—á–∏—Å—Ç–∫–∏

```python
data['clean_text'] = data['text'].progress_apply(lemmatize_spacy)
```

```python
data['clean_text'] = data['clean_text'].progress_apply(clear_text)
```

```python
data.sample(7)
```

```python
text = ' '.join(data.query('toxic==0')['clean_text'])

wordcloud = WordCloud().generate(text)
plt.figure(figsize=(9,5))
plt.imshow(wordcloud)
plt.title('–û–±–ª–∞–∫–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤')
plt.axis('off');
```

–ö–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ –æ–±–ª–∞–∫–∞ —á–∞—â–µ –≤—Å–µ–≥–æ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è —Å–ª–æ–≤–∞: article(—Å—Ç–∞—Ç—å—è),think(–¥—É–º–∞—Ç—å),make(–¥–µ–ª–∞—Ç—å),edit(—Ä–µ–¥–∞–∫—Ç–∏—Ä–æ–≤–∞—Ç—å),page(—Å—Ç—Ä–∞–Ω–∏—Ü–∞) –∏ –µ—â–µ –º–Ω–æ–≥–æ –¥—Ä—É–≥–∏—Ö,–∫–æ—Ç–æ—Ä—ã–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –æ–ø—Ä–µ–¥–µ–ª—è—é—Ç —Ç–µ–º–∞—Ç–∏–∫—É –Ω–∞—à–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è –∏ –Ω–µ –Ω–æ—Å—è—Ç –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä.

```python
text = ' '.join(data.query('toxic==1')['clean_text'])

wordcloud = WordCloud().generate(text)
plt.figure(figsize=(9,5))
plt.imshow(wordcloud)
plt.title('–û–±–ª–∞–∫–æ –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤')
plt.axis('off');
```

–ò–∑ –¥–∞–Ω–Ω–æ–≥–æ –æ–±–ª–∞–∫–∞ –≤–∏–¥–Ω–æ, —á—Ç–æ –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã–µ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ —Ä–∞–∑–º–µ—á–µ–Ω—ã –≤–µ—Ä–Ω–æ –∏ –∫–∞–∫–æ–µ —Å–ª–æ–≤–æ —è–≤–ª—è–µ—Ç—Å—è —Å–∞–º—ã–º —É–ø–æ—Ç—Ä–µ–±–ª—è–µ–º—ã–º )))

```python
# –†–∞–∑–æ–±—å–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É —Å —É—á–µ—Ç–æ–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞

train, test = train_test_split(data, test_size=0.25,random_state=12345,stratify=data['toxic'])
```

```python
print('–û—Ç–Ω–æ—à–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ:',train['toxic'].value_counts()[1]/train['toxic'].value_counts()[0])
print('–û—Ç–Ω–æ—à–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:',test['toxic'].value_counts()[1]/test['toxic'].value_counts()[0])
```

```python
plt.figure(figsize=(6, 6))
plt.subplot(211)
train['toxic'].hist()
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ train')

plt.subplot(212)
test['toxic'].hist()
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ test')


```

```python
# –°–æ–∑–¥–∞–ª–∏ –∫–æ—Ä–ø—É—Å—ã –∏ –æ—Ç–¥–µ–ª–∏–ª–∏ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –≤ –æ–±–µ–∏—Ö –≤—ã–±–æ—Ä–∫–∞—Ö

corpus_train = list(train['clean_text']) # —Å–æ–∑–¥–∞–ª–∏ –∫–æ—Ä–ø—É—Å –∏–∑ —Ç—Ä–µ–π–Ω–∞
train_y = train['toxic']

corpus_test = list(test['clean_text']) # —Å–æ–∑–¥–∞–ª–∏ –∫–æ—Ä–ø—É—Å –∏–∑ —Ç–µ—Å—Ç–∞
test_y = test['toxic']
```

# –û–±—É—á–µ–Ω–∏–µ –∏ –≤—ã–±–æ—Ä –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ 

```python
stopwords = set(nltk_stopwords.words('english'))# –ø—Ä–æ–≥—Ä—É–∑–∏–ª–∏ —Å—Ç–æ–ø-—Å–ª–æ–≤–∞ –Ω–∞ –∞–Ω–≥–ª–∏–π—Å–∫–æ–º
```

–î–ª—è –≤—ã–±–æ—Ä–∞ –º–æ–¥–µ–ª–∏ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å pipline

```python
pipe = Pipeline(
    [
        ("vect", TfidfVectorizer(stop_words = stopwords,ngram_range=(1,2))),
        ("clf", LogisticRegression()),
    ]
)
```

–°–Ω–∞—á–∞–ª–∞ –±–µ–∑ –ø–æ–¥–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≤—ã–±–µ—Ä–µ–º –º–æ–¥–µ–ª—å. –ì–¥–µ –µ—Å—Ç—å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å,—É–∫–∞–∑—ã–≤–∞–µ–º –Ω–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç—å —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ.

```python
%%time

param_grid = [
    {
        "clf": 
        [
        LogisticRegression(random_state=12345,class_weight='balanced'),
        SGDClassifier(random_state=12345,class_weight='balanced'),
        LinearSVC(class_weight='balanced', random_state=12345),
        ],
    }
]
grid = GridSearchCV(pipe, param_grid=param_grid, scoring='f1', cv=5, n_jobs=-1)
grid.fit(corpus_train, train_y)
```

```python
print(f'–õ—É—á—à–∞—è F1:{grid.best_score_:.4f}')
```

```python
print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å:')
grid.best_estimator_.named_steps.clf
```

–ü–æ–¥–±–µ—Ä–µ–º –¥–ª—è –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã

```python
param_grid_best_model = [
    {
        "clf": [LinearSVC(class_weight='balanced', random_state=12345)],
        "clf__C" : (0.1,0.56,2.52,3.16),
       
    },
]
```

```python
%%time
grid_best_model = GridSearchCV(pipe, param_grid=param_grid_best_model, scoring="f1", cv=5,n_jobs=-1)
grid_best_model.fit(corpus_train, train_y)
```

```python
print(f'–ú–µ—Ç—Ä–∏–∫–∞ F1 –Ω–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ —Å –ø–æ–¥–æ–±—Ä–∞–Ω–Ω—ã–º–∏ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏: {grid_best_model.best_score_:.4f}')
```

```python
print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å')
grid_best_model.best_estimator_.named_steps.clf
```

## –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏

```python

print(f'F1 –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º –Ω–∞–±–æ—Ä–µ –¥–∞–Ω–Ω—ã—Ö: {f1_score(test_y, grid_best_model.predict(corpus_test)):.4f}')
```

  - –æ—á–∏—Å—Ç–∏–ª–∞ –æ—Ç –ª–∏—à–Ω–∏—Ö –ø—Ä–æ–±–µ–ª–æ–≤ –∏ –∑–Ω–∞–∫–æ–≤
  - —Ä–∞–∑–±–∏–ª–∞ –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ —Å —É—á–µ—Ç–æ–º –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ –∫–ª–∞—Å—Å–æ–≤

  - –¥–ª—è –Ω–∞—á–∞–ª–∞ –±–µ–∑ –ø–æ–¥–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–º–µ—Ç—Ä–æ–≤ –æ–ø—Ä–µ–¥–µ–ª–∏–ª–∞ –ª—É—á—à—É—é –º–æ–¥–µ–ª—å, —Å —É–¥–∞–ª–µ–Ω–∏–µ–º —Å—Ç–æ–ø-—Å–ª–æ–≤. –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ª—É—á—à–µ–π –º–æ–¥–µ–ª—å—é –æ–ø—Ä–µ–¥–µ–ª–∏–ª–∞—Å—å **LinearSVC(class_weight='balanced', random_state=12345)**
  - –¥–ª—è –Ω–µ–µ –ø–æ–¥–æ–±—Ä–∞–ª–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏ C.

**–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å :** LinearSVC(C=3.16, class_weight='balanced', random_state=12345)

**F1 –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º –Ω–∞–±–æ—Ä–µ —Å–æ—Å—Ç–∞–≤–∏–ª–∞ :** 0.7895
