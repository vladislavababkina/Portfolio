# Определяющие успешность игры закономерности

Имеем исторические данные до 2016 года о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation)\
**Цель:** выявить определяющие успешность игры закономерности.\
**Описание данных:**
-  Name — название игры
-  Platform — платформа
-  Year_of_Release — год выпуска
-  Genre — жанр игры
-  NA_sales — продажи в Северной Америке (миллионы проданных копий)
-  EU_sales — продажи в Европе (миллионы проданных копий)
-  JP_sales — продажи в Японии (миллионы проданных копий)
-  Other_sales — продажи в других странах (миллионы проданных копий)
-  Critic_Score — оценка критиков (максимум 100)
-  User_Score — оценка пользователей (максимум 10)
-  Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Поскольку оценки пользователей и критиков в нашем случае никак не связаны с рейтингом, то заменю их на -1, чтобы не потерять такой объем данных, но видеть что изначельно тут были пропуски. В рейтинге пропуски заменю на RP («Rating Pending») — Рейтинг ожидается.

```python
data['critic_score'] = data['critic_score'].fillna(-1)
data['user_score'] = data['user_score'].fillna(-1)
data['rating'] = data['rating'].fillna('RP')
```

TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято.

```python
data.query('user_score == "tbd"')
```

Заменю tbd на -1 , так как 0  в рейтинге это уже оценный продукт, а  оценка в данном случае еще не определена. Так же нет связи оценок и рейтинга, поскольку рейтинг от организации ESRB связан с возрастной группой, а не оценками пользователей или критиков.

```python
data.loc[data['user_score']=="tbd",'user_score'] = -1
```

Предполагаю,что такое количество пропусков имеет технический характер, например при слиянии таблиц. Что касается пропусков в рейтинге ESRB, то вероятнее всего у части игр он просто отстутвует, так как ESRB-рейтинговая система развлекательного ПО рассчитаная на рынок США и Канады.Соотвественно, для игр, целевой аудиторией которых являются Европа и Япония, этот рейтинг мало актуален.

```python
data['year_of_release'] = data['year_of_release'].astype(int)
data['critic_score'] = data['critic_score'].astype(int)
data['user_score'] = data['user_score'].astype(float)
```

```python
critic_score_sales = data_last_3_years.query(('platform == "PS4"') and ('critic_score != -1'))
critic_score_sales.plot(y='general_sales',x='critic_score',kind='scatter', alpha = 0.3)
plt.title('Зависимость продаж и оценок критиков')
plt.ylabel('объемы продаж')
plt.xlabel('оценка критиков')
user_score_sales = data_last_3_years.query(('platform == "PS4"') and ('user_score != -1'))
user_score_sales.plot(y='general_sales',x='user_score',kind='scatter', alpha = 0.3,color='r')
plt.title('Зависимость продаж и оценок пользователей')
plt.ylabel('объемы продаж')
plt.xlabel('оценка пользователей')
print('Корреляция между продажами и отзывами критиков равна',critic_score_sales['general_sales'].corr(critic_score_sales['critic_score']))
print('Корреляция между продажами и отзывами пользователей равна',user_score_sales['general_sales'].corr(user_score_sales['user_score']))
```

Зависимость между продажами и оценками критиков положительная, но очень незначительная.А вот между продажами и оценками пользователей отрицательная и слишком близка к нулю.\
Таким образом оценки пользователей не влияют на продажи, а оценки критиков очень незначительно,возможно часть пользователей все же доверяет их мнению.

```python
platform_name = data_last_3_years['platform'].unique()
for name in platform_name:
    platform_unique = data_last_3_years.loc[data_last_3_years['platform']== name]
    print( 'Корреляция между продажами и оценкой критиков', name ,'равна', platform_unique['general_sales'].corr(platform_unique['critic_score']))
    print( 'Корреляция между продажами и оценкой пользователей', name ,'равна', platform_unique['general_sales'].corr(platform_unique['user_score']))
    print()
```

Как видно, если для самой популярной платформы нет связи между продажами и оценками, так и для остальных платформ наблюдаем похожую ситуацию, за исключением Wii, где наблюдается положительная зависимость продаж от оценки пользователей.

##   Портрет пользователя каждого региона

```python
def top_5_platform(column):
    ### функция создает сводную таблицу топ-5 платформ по продажам ###
    table = (
        data_last_3_years.pivot_table(index = 'platform', values = column, aggfunc = 'sum')
        .sort_values(column, ascending = False).reset_index().head(5)
    )
    table['share_of_sales'] = (table[column]/table[column].sum()*100).round(2)
    return table
```

```python
na = top_5_platform('na_sales')

eu = top_5_platform('eu_sales')

jp = top_5_platform('jp_sales')
```

```python
display(na,eu,jp)
```

```python
all_region = pd.merge(na, eu, on = 'platform',how='outer')
all_region = pd.merge(all_region, jp, on = 'platform',how='outer')
all_region.columns=['platform','na_sales','%_na','eu_sales','%_eu','jp_sales','%_jp']
all_region
```

Видим,что 3 платформы популярны у пользователей всех трех регионов : PS4, 3DS, PS3.\
Сильно отличаются пользователи Японии отдавая предпочтение местной Nintendo.\
Так в Северной Америк 52,4% отдают предпочтение платформам PS4 и XOne,в Европе  62,55% предпочтение платформам PS4 и PS3.В Японии же большую долю занимают платформы 3DS и PS3, составляя 66,73% от общего числа пользователей в регионе.\
Пользователи Европы и Северной Америки сходятся в предпочтениях топ-5 платформ,а в Японии XOne и X360 предпочитают игры на платформах PSV и WiiU

```python
def top_5_genre(column):
    ### функция создает сводную таблицу топ-5 жанров по продажам ###
    table = (
        data_last_3_years.pivot_table(index = 'genre', values = column, aggfunc = 'sum')
        .sort_values(column, ascending = False).reset_index().head(5)
    )
    table['share_of_sales'] = (table[column]/table[column].sum()*100).round(2)
    return table

```

```python
na_genre = top_5_genre('na_sales')
eu_genre = top_5_genre('eu_sales')
jp_genre = top_5_genre('jp_sales')
display(na_genre,eu_genre,jp_genre)
```

И снова выбор пользователей Японии отличается от Европы и Северной Америки.\
Так 80% пользователей в Европе и Америке предпочитают игры в жанрах Action,Shooter и Sports, а в Японии те же 80% Role-Playing и Action.

Рассмотрим предпочтение игр по рейтингам, уберем игры с рейтингом RP, поскольку ранее мы им заполнили пропуски и неопределенный рейтинг для нас в принципе не интересен.

Организации ESRB (англ. Entertainment Software Rating Board) - ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.Ниже краткое описание:
-  «EC» («Early childhood») — Для детей младшего возраста
-  «E» («Everyone») — Для всех
-  «E10+» («Everyone 10 and older») — Для всех от 10 лет и старше
-  «T» («Teen») — Подросткам
-  «M» («Mature») — Для взрослых
-  «AO» («Adults Only 18+») — Только для взрослых
-  «RP» («Rating Pending») — Рейтинг ожидается

```python
def top_5_rating(column):
    ### функция создает сводную таблицу топ-5 возрастных рейтингов по продажам ###
    data_no_rp = data_last_3_years.query('rating != "RP"')
    table = (
        data_no_rp.pivot_table(index = 'rating', values = column, aggfunc = 'sum')
        .sort_values(column, ascending = False).reset_index().head(5)
    )
    table['share_of_sales'] = (table[column]/table[column].sum()*100).round(2)
    return table
```

```python
na_rating = top_5_rating('na_sales')
eu_rating = top_5_rating('eu_sales')
jp_rating = top_5_rating('jp_sales')
display(na_rating,eu_rating,jp_rating)
```

Предпочтения Европы и Северной Америки схожи, первые строчки занимают игры с рейтингом "Для взрослых", в Японии на первом месте игры для подростков, однако не стоит забывать о том, что для Японии рейтинг ESRB мало актуален, поэтому по данному региону нет возможности сделать достоверный вывод о предпочтении возрастной категории игр.

##   Проверка гипотезы

Проверим гипотезу о равенстве реитинга платформ Xbox One и PC. Для этого выдвинем нулевую гипотезу о равенстве и альтернативную о не равенстве.\
**H₀: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые(равны)**\
**H₁: Средние пользовательские рейтинги платформ Xbox One и PC разные(не равны)**

Поскольку нам необходимо проверить равенство средних в двух выборках применим критерий Стьюдента.

```python
xone_rating = data_last_3_years.loc[(data_last_3_years['platform']=="XOne")&
                               (data_last_3_years['user_score'] != -1),'user_score']
pc_rating = data_last_3_years.loc[(data_last_3_years['platform']=="PC")&
                               (data_last_3_years['user_score'] != -1),'user_score']
print(xone_rating.shape[0])
print(pc_rating.shape[0])
```

Для проверки применяем t- тест:
-  Генеральные совокупности не зависят друг от друга;
-  Выборочные средние нормально распределены,согласно ЦПТ, выборка больше 30;
-  Выборки достаточно велики,поэтому предположу что дисперсии равны, тест редко ошибается даже если дисперсии на самом деле не равны, при большом количестве значений.

```python
alpha = .05
results = st.ttest_ind(xone_rating, pc_rating)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вывод: вероятность случайно получить такое или большее различие равна почти 15 %. Это слишком большая вероятность, чтобы делать вывод о значимом различии между пользовательскими рейтингами платформ Xbox One и PC.

Проверим гипотезу о равенстве средних пользовательских рейтингов жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования»).\
**H₀: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») равны.**\
**H₁: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.**

```python
action_rating = (data_last_3_years.loc[(data_last_3_years['genre']=="Action") 
                                       & (data_last_3_years['user_score'] != -1),'user_score'])
spots_rating = (data_last_3_years.loc[(data_last_3_years['genre']=="Sports") 
                                      & (data_last_3_years['user_score'] != -1),'user_score'])
print(action_rating.shape[0])
print(spots_rating.shape[0])
```

Выборки сильно отличаются по размеру, применим параметр equal_var = False, все остальные условия подходят для применения t-теста.

```python
alpha = .05
results = st.ttest_ind(action_rating, spots_rating, equal_var = False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Делаем вывод о значимом различии между пользовательскими рейтингами жанров Action и Sports.
